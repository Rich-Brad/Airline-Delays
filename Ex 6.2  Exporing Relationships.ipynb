# Task 3 Create a correlaton matrix heatmap (colored)

```python
plt.matshow(df_air_dlay_1.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(15, 15)) # figure size 
plt.matshow(df_air_dlay_1.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_air_dlay_1.shape[1]), df_air_dlay_1.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(df_air_dlay_1.shape[1]), df_air_dlay_1.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Airline Delay Correlation Matrix', fontsize=14) # add title
```

```python
# Create a correlation heatmap of 'df_air_dlay_1'

plt.figure(figsize = (10,10))
sns.heatmap(df_air_dlay_1.corr(), annot = True)
```

# Task 4 Create a scatterplot for the variables with the strongest correlations and examine the nature of their relationship.

```python
# Create a scatterplot

sns.scatterplot(x = 'number_arriving_flights', y = 'number_flights_delay_greater_than_15min', data = df_air_dlay_1)
```

# The number_arriving_flights is all of the aircraft arriving at the airports within this data set.  The number_flights_delay_greater_than_15min is a subset of the number_arriving_flights at those airports. There is a stong relationship when there is an increase of number_arriving_flights there is an increase of number_flights_delay_greater_than_15min. This strong trend is indicated by the steep lower left cluster to upper right corner of the scatterplot.
