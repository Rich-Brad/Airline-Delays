# 5. State my hypothesis

# “If the arriving flights experience heavy air traffic, then the arriving flights will more likely be delayed 15 minutes or longer than flights that do not experience heavy air traffic."

# 6. Reshape the variables into NumPy arrays with x being the independent variable and y being the dependent variable.

```python
X = df_air_dlay['number_flights_heavy_air_traffic_delay'].values.reshape(-1,1)
y = df_air_dlay['number_flights_delay_greater_than_15min'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

# 7. Split the data into two sets: a training set and a test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

# 8. Run a linear regression on the data

```python
# Training set predict values of y using X.

regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
y_predicted = regression.predict(X_test)
```

```python
# Create a prediction for y on the test set

y_predicted = regression.predict(X_test)
```

# 9. Create a plot that shows the regression line on the test set.

```python
plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('Heavy air traffic to flight delays 15 minutes or greater (Test set)')
plot_test.xlabel('Heavy air traffic delays in minutes')
plot_test.ylabel('Flight delays 15 minutes or greater from their scheduled arrival time')
plot_test.show()
```

# Task 10. Write your own interpretation of how well the line appears to fit the data in markdown cell. 

# The regression line does not fit perfectly or cover all the dots.  There are many points that show when there is heavy air traffic there are flight delays that are 15 minutes or greater. 

# The slope of the line is positive, see  how the regression line starts from the bottom left and rises towards the top right. This suggests that there is a positive relationship between heavy air traffic and flight delays.  

# The coefficent is also positive. As 'X' or 'Heavy air traffice delays in minutes' rises, the 'y' or 'Flight delays 15 minutes or greater' also rises.  This indicates a positive relationship, so the coefficients positive as well. 

# Task 11. Check the model performance statistics MSE and R2 score.

```python
# Create objects that contain the model summery statistics

rmse = mean_squared_error(y_test, y_predicted) # This is the mean squared error
r2 = r2_score(y_test, y_predicted) # This is the R2 score. 
```

```python
# Print the model summary statistics to evaluate the performance of the model

print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```
